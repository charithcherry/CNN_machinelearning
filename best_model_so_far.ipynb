```
datadir='/content/train'
os.listdir(datadir)
```

```
K.clear_session()

n_classes = 11
img_width, img_height = 299, 299
batch_size=16
nb_train_samples = 3454
nb_validation_samples =1477 
```

```
datagen = ImageDataGenerator(validation_split=0.3,
    rescale=1. / 255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True)
train_generator = datagen.flow_from_directory(
    datadir,
    subset='training',
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical')
val_generator = datagen.flow_from_directory(
    datadir,
    subset='validation',
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical')

```

```
inception = InceptionV3(weights='imagenet', include_top=False)
x = inception.output
x = GlobalAveragePooling2D()(x)
x = Dense(128,activation='relu')(x)
x = Dropout(0.2)(x)

predictions = Dense(11,kernel_regularizer=regularizers.l2(0.005), activation='softmax')(x)

model = Model(inputs=inception.input, outputs=predictions)
model.compile(optimizer=SGD(lr=0.0001, momentum=0.9), loss='categorical_crossentropy', metrics=['accuracy'])
checkpointer = ModelCheckpoint(filepath='best_model_11class.hdf5', verbose=1, save_best_only=True)
csv_logger = CSVLogger('history_11class.log')

history_11class = model.fit_generator(train_generator,
                    steps_per_epoch = nb_train_samples // batch_size,
                    validation_data=val_generator,
                    validation_steps=nb_validation_samples // batch_size,
                    epochs=30,
                    verbose=1,
                    callbacks=[csv_logger, checkpointer])
model.save('model_trained_11class.hdf5')
```

```
def plot_accuracy(history,title):
    plt.title(title)
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.ylabel('accuracy')
    plt.xlabel('epoch')
    plt.legend(['train_accuracy', 'validation_accuracy'], loc='best')
    plt.show()
def plot_loss(history,title):
    plt.title(title)
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.ylabel('loss')
    plt.xlabel('epoch')
    plt.legend(['train_loss', 'validation_loss'], loc='best')
    plt.show()
```

```
plot_accuracy(history_11class,'FOOD11-Inceptionv3')
plot_loss(history_11class,'FOOD11-Inceptionv3')
```

```

def predict_class(model, folder_path, show = True):
 for img in os.listdir(folder_path):
    img_name.append(img)
    img = image.load_img(folder_path+'/'+img,grayscale=False, color_mode='rgb',target_size=(299, 299))
    img = image.img_to_array(img)                    
    img = np.expand_dims(img, axis=0)         
    img /= 255.                                      

    pred = model.predict(img)
    index = np.argmax(pred)
    indices.append(index)
```
